```python
import random
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
sns.set()
plt.rcParams["figure.figsize"] = (16,9)
x_axis = np.arange(0, 10, 0.01)
mu = 5
va = 2
```

```python
x=[]
for i in range(5000):
  r = random.gauss(mu, va)
  x.append(r)
```

```python
plt.hist(x, density=True, bins= 30, label='Obtido')
plt.xlim(0, 10)
plt.plot(x_axis, norm.pdf(x_axis,mu,va), label='Teórico')
plt.title('Distruibuição de N=5000 do gerador gaussiano de números aleatórios, Variancia=4 e Média 5')
plt.ylabel('Densidade')
plt.xlabel('Valor')
plt.axvline(5, color ='black', label='Média = 5')
plt.legend()
```

```python
r=[]

for i in range(500000):
  x = random.random() *8
  y = random.random()
  if y < np.exp(-x):
    r.append(x)
xx = np.linspace(0, 8, 1000)


plt.title('Distruibuição de N=500000 do gerador exponencial ')
plt.ylabel('Densidade')
plt.xlabel('Valor')
plt.hist(r, density=True, label='Obtido')
plt.plot(xx, np.exp(-xx), label='Teórico')
#plt.yscale('log')
```

```python
plt.title('Distruibuição de N=500000 do gerador exponencial ')
plt.ylabel('Log Densidade')
plt.xlabel('Valor')
plt.hist(r, density=True, label='Obtido')
plt.plot(xx, np.exp(-xx), label='Teórico')
plt.yscale('log')
```
