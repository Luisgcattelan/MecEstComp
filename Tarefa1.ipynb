```python
import numpy as np
from matplotlib import pyplot as plt
import random
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (16,9)
```

```python
n=range(1, 1000)
random.seed(42)


medk=0
medk2=0
lisk=[]
lisk2=[]
avek=0
avek2=0
erro=0
for i in n:
  k=random.randint(0, 1)
  medk= medk+k
  medk2 = medk2 + k^2
  avek= medk/ (i)
  avek2= medk2/ (i)
  if i==1:
    erro=np.nan
  else:
      erro= np.sqrt((avek2 - (avek*avek))/(i-1))  

  lisk.append(avek)
  lisk2.append(erro)

```

```python
plt.errorbar(n,lisk, yerr=lisk2, ecolor='black')  
plt.title('Média e barra de erro de 1000 moedas' )
plt.axhline(0.5, c='black')
plt.xlabel('Número de passos')
plt.ylabel('Média')
plt.xlim(0,1000)
plt.ylim(0,1)
```

```python
plt.title('Desvio padrão de 1000 moedas' )
plt.xlabel('Número de passos')
plt.ylabel('Desvio padrão')
plt.scatter(n, lisk2, c='black')
plt.xlim(2,)
```
