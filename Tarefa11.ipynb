```python
## Euler's method
from numpy import zeros, linspace, pi, cos, array
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
omega = 1
m=1
N=500
dt=0.08
t = linspace(0, N*dt, N+1)

u = zeros(N+1)
vi = zeros(N+1)

X_0 = 2
u[0] = X_0
vi[0] = 0

for n in range(N):
    u[n+1] = u[n] + dt*vi[n]
    vi[n+1] = vi[n] - dt*omega**2*u[n]


uv = zeros(N+1)
viv = zeros(N+1)

X_0 = 2
uv[0] = X_0
viv[0] = 0

for n in range(N):
    v_next = viv[n] - 0.5*omega**2*uv[n]*dt
    uv[n+1]= uv[n]+ dt*v_next
    viv[n+1] = v_next - 0.5*omega**2*uv[n+1]*dt



plt.plot(t, u, 'b-', label = 'Euler')
         
plt.plot(t, X_0*cos(omega*t), 'r--', label = 'Análitico')

plt.plot(t, uv, 'y--', label = 'Verlet')
plt.legend()
plt.title('Posição por tempo')
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

```python
def energy(u,v):

  E = (v**2)/2 + (u**2)/2
  dE = (E-E[0])/E[0]
  return dE


ee = energy(u,vi) 
plt.plot(t,ee)
plt.plot(t,eev)


eev = energy(uv,viv) 
```

```python
fig, axs = plt.subplots(3,1, figsize=(6,16))
axs[0].plot(t, u, 'b-', label = 'Euler')       
axs[0].plot(t, X_0*cos(omega*t), 'r--', label = 'Análitico')
axs[0].plot(t, uv, 'y--', label = 'Verlet')
axs[0].legend()
axs[0].set(title='Posição por tempo', xlabel='t', ylabel='x')

axs[1].plot(t, vi, 'b-', label = 'Euler')       
axs[1].plot(t, viv, 'y--', label = 'Verlet')
axs[1].legend()
axs[1].set(title='Velocidade por tempo', xlabel='t', ylabel='v')


axs[2].plot(t, ee, 'b-', label = 'Euler')       
axs[2].plot(t, eev, 'y--', label = 'Verlet')
axs[2].legend()
axs[2].set(title='$\Delta E$ por tempo', xlabel='t', ylabel='$\Delta E(t)/E_0$')


plt.savefig('Tarefa11.pdf')
```
